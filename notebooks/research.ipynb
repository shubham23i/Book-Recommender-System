```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
common_args = {
    "sep": ";",
    "encoding": "latin-1",
    "engine": "python",
    "on_bad_lines": "skip",
    "quoting": 3  
}

books = pd.read_csv("books_data/books.csv", **common_args)
users = pd.read_csv("books_data/users.csv", **common_args)
ratings = pd.read_csv("books_data/ratings.csv", **common_args)
```

```python
books=books[['"ISBN"', '"Book-Title"', '"Book-Author"', '"Year-Of-Publication"',
       '"Publisher"','"Image-URL-L"']]
```

```python
users.columns
```

```python
books.columns = books.columns.str.replace('"', '').str.strip()
users.columns = users.columns.str.replace('"', '').str.strip()
ratings.columns = ratings.columns.str.replace('"', '').str.strip()
```

```python
books.rename(columns={
    'ISBN':'isbn',
    'Book-Title':'title',
    'Book-Author':'author',
    'Year_Of_Publication':'year',
    'publisher':'publisher',
    'Image-URL-L':'image_url',
},inplace=True)

```

```python
ratings.columns
```

```python
users.rename(columns={
    'User-ID':'user_id',
    'Location':'location',
    'Age':'age'
},inplace=True)
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'ISBN':'isbn',
    'Book-Rating':'book_rating'
},inplace=True)
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200
x.value_counts()
```

```python
y=x[x].index
y
```

```python
rating=ratings[ratings['user_id'].isin(y)]
rating.head()
```

```python
ratings_of_books=pd.merge(books,ratings,on='isbn')
```

```python
ratings_of_books.head(2)
```

```python
number_rating=ratings_of_books.groupby('title')['book_rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'book_rating':'num_of_ratings'},inplace=True)
```

```python
final_rating=pd.merge(number_rating,ratings_of_books,on='title')
```
