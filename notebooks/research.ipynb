```python
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.columns = books.columns.str.replace('"', '').str.strip()
users.columns = users.columns.str.replace('"', '').str.strip()
ratings.columns = ratings.columns.str.replace('"', '').str.strip()
```

```python
books.rename(columns={
    'ISBN':'isbn',
    'Book-Title':'title',
    'Book-Author':'author',
    'Year_Of_Publication':'year',
    'publisher':'publisher',
    'Image-URL-L':'image_url',
},inplace=True)

```

```python
users.rename(columns={
    'User-ID':'user_id',
    'Location':'location',
    'Age':'age'
},inplace=True)
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'ISBN':'isbn',
    'Book-Rating':'rating'
},inplace=True)
```

```python
ratings_of_books=pd.merge(books,ratings,on='isbn')
```

```python
ratings_of_books.head(2)
```

```python
number_rating=ratings_of_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_ratings'},inplace=True)
```

```python
final_rating=pd.merge(number_rating,ratings_of_books,on='title')
```

```python
final_rating.head(2)
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.columns
```

```python

final_rating.isna().sum()
```

```python
print(final_rating['rating'].dtype)
final_rating['rating'] = pd.to_numeric(final_rating['rating'], errors='coerce')
final_rating = final_rating[final_rating['num_of_ratings'] >= 50]
```

```python
print(final_rating.shape)
print(final_rating['rating'].isna().sum())
print(final_rating.head())
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=10)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name='A Case of Need'
recommend_book(book_name)
```
